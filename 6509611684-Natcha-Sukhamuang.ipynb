```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/data_updated.csv', delimiter=',')
```

```python
df.head()
```

```python
data = pd.DataFrame(df)
```

```python
X = list(data.iloc[: , 0])
Y = list(data.iloc[: , 1])
```

```python
#Plot the data using bar() method
plt.bar (X, Y, color = 'g')
plt.title("Students over 11 Years")
plt.xlabel("Years")
plt.ylabel("Number of Students")
```

```python
#Show the plot
plt.show
```
