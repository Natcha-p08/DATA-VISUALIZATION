```python
# นับจำนวนแต่ละเพศ
gender_counts = df['เพศ'].value_counts()

# กำหนดสีที่ต้องการ
colors = ['#f32f65', '#1b82c5', '#773b87','#f89c25' , '#5ca2d4' , '#98c32b', '#66addb' ]


# Disable LaTeX rendering
plt.rcParams["text.usetex"] = False

# สร้าง Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('เพศ', pad=20, fontweight='bold')  # กำหนดชื่อ title และให้ตัวหนา (bold)
plt.axis('equal')  # ทำให้ Pie Chart เป็นวงกลมที่แท้จริง
plt.show()
```
